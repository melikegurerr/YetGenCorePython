# Problem 3

3 öğrenciden oluşan bir öğrenci not sözlüğü oluşturun. Bu sözlükte öğrencilerin notları value olarak bir listede toplansın.

Kullanıcıya hangi öğrencinin notlarını görmek istediğini sorun. Öğrencinin notu görüntülendiğinde program sonunda şöyle bir çıktı elde etmelisiniz:

```
Lütfen notlarını görmek istediğiniz öğrencinin adını girin: Mehmet

Mehmet isimli öğrencinin      1.Sınav Notu:72
                              2.Sınav Notu:66
                              3.Sınav Notu:48
Not Ortalaması: 62.0


````python
ogrencinotlari = {
    "Melike": [72, 66, 48],
    "Elif": [85, 90, 78],
    "Levent": [60, 70, 75]
}

ogrenciadi = input("Lütfen notlarını görmek istediğiniz öğrencinin adını girin: ")

if ogrenciadi in ogrencinotlari:
    notlar = ogrencinotlari[ogrenciadi]
    not_sayisi = len(notlar)
    not_ortalamasi = sum(notlar) / not_sayisi
    
    print(f"\n{ogrenciadi} isimli öğrencinin")
    for i in range(not_sayisi):
        print(f"{i+1}.Sınav Notu: {notlar[i]}")
    print(f"Not Ortalaması: {not_ortalamasi}")
else:
    print("Girdiğiniz öğrenci adı bulunamadı.")
````
